# 파이썬 예외 처리

## 기본 예외 처리

```python
# 기본 예외 처리
try:
    # 코드 블록
    result = 10 / 0
    print("결과:", result)
except Exception as e:
    print("0으로 나눌 수 없습니다:", e)
```

## 여러 개의 예외 처리

```python
# 여러 개의 예외 처리: 먼저 발생한 예외에서 중단 - "abc"를 "123"으로 변경하면 ValueError는 발생하지 않음
try:
    # 코드 블록
    value = int("2400")  # ValueError 발생
    # value = int("123")  # ValueError 발생하지 않음
    result = '1' + value    # TypeError 발생
    print("변환된 값:", value)
except ValueError as e:
    print("잘못된 값입니다:", e)
except TypeError as e:
    print("잘못된 타입입니다:", e)
```

## 예외 여부와 관계 없이 마감 처리 - `finally` 

```python
# finally 블록: 예외 여부와 관계 없이 마감 처리
try:
    result = 10 / 0
    #result = 10 / 2
    print("결과:", result)
except ZeroDivisionError as e:
    print("0으로 나눌 수 없습니다:", e)
finally:
    print("계산을 마무리 합니다.")
```
