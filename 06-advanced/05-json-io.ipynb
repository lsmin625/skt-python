# CSV와 JSON 파일 처리

## CSV 파일 읽고 쓰기

```python
# CSV 파일 쓰기 예시
import csv
import random

# 랜덤하게 CSV 데이터를 생성하는 함수 - n개를 매개 변수로 받음
def generate_random_data(n):
    names = ['홍길동', '임꺽정', '연놀부', '심청', '성춘향']
    cities = ['서울', '부산', '대구', '인천', '광주']
    data = []
    for _ in range(n):
        name = random.choice(names)
        age = random.randint(18, 60)
        city = random.choice(cities)
        data.append([name, age, city])
    return data

NUM_OF_ROWS = 20

with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    for row in generate_random_data(NUM_OF_ROWS):
        writer.writerow(row)
```

```python
# CSV 파일 읽기 예시
with open('data.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

## JSON 파일 읽고 쓰기

```python
# json 모듈을 사용한 JSON 파일 읽고 쓰기 예시
import json

data = {
    'employees': [
        {'name': '홍길동', 'age': 30, 'city': '서울'},
        {'name': '임꺽정', 'age': 35, 'city': '부산'},
        {'name': '연놀부', 'age': 28, 'city': '대구'},
        {'name': '심청', 'age': 25, 'city': '인천'},
        {'name': '성춘향', 'age': 32, 'city': '광주'}
    ]
}

# JSON 파일로 데이터 쓰기
with open('data.json', 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)
```

```python
# JSON 파일에서 데이터 읽기
with open('data.json', 'r', encoding='utf-8') as json_file:
    loaded_data = json.load(json_file)
    print(loaded_data['employees'][0])
```
