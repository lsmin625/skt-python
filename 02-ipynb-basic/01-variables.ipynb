# 파이썬 자료구조

## 파이썬 변수 유형

| 변수 유형 | 설명 | 예시 |
|---------|------|------|
| int | 정수형 숫자 | `age = 25` |
| float | 실수형 숫자 | `pi = 3.14` |
| str | 문자열 | `name = "홍길동"` |
| bool | 불리언 (참/거짓) | `is_active = True` |
| list | 순서가 있는 변경 가능한 컬렉션 | `numbers = [1, 2, 3]` |
| tuple | 순서가 있는 변경 불가능한 컬렉션 | `point = (10, 20)` |
| dict | 키-값 쌍의 컬렉션 | `person = {"name": "김철수", "age": 30}` |
| set | 중복 없는 순서 없는 컬렉션 | `unique = {1, 2, 3}` |
