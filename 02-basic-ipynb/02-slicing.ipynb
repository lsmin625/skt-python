# 문자열 처리

## 문자열 슬라이싱
문자열 슬라이싱은 문자열의 특정 부분을 추출하는 방법
- 슬라이싱 구문은 `문자열[시작:끝:단계]` 형식을 따르며, 시작 인덱스부터 끝 인덱스 전까지의 부분 문자열을 반환
- 단계는 선택적으로 지정할 수 있으며, 기본값은 1.

```python
# 문자열 슬라이싱 예시
text = "Hello, Python!"
substring = text[0:5]  # 'Hello'
print(f"슬라이싱된 문자열: {substring}")

# 문자열 뒤집기
reversed_text = text[::-1]
print(f"뒤집힌 문자열: {reversed_text}")
```
