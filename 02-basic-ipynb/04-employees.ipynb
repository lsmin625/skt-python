# 리스트 필터링

```python
employees = [
{"name": "Alice", "department": "Engineering", "age": 30, "salary": 85000},
{"name": "Bob", "department": "Marketing", "age": 25, "salary": 60000},
{"name": "Charlie", "department": "Engineering", "age": 35, "salary": 95000},
{"name": "David", "department": "HR", "age": 45, "salary": 70000},
{"name": "Eve", "department": "Engineering", "age": 28, "salary": 78000},
]
```

```python
# 부서가 "Engineering"이고 salary >= 80000인 직원들의 이름만 리스트로 출력하세요
result = [emp["name"] for emp in employees  if emp["department"] == "Engineering" and emp["salary"] >= 80000]
print(result)
```

```python
# 30세 이상인 직원의 이름과 부서를 튜플 (name, department) 형태로 리스트로 출력하세요.
result = [
    (emp["name"], emp["department"])
    for emp in employees
    if emp["age"] >= 30
]

print(result)
```

```python
# 급여 기준으로 직원 리스트를 salary 내림차순으로 정렬하고, 상위 3명의 이름과 급여를 출력하세요.
# 급여 기준 내림차순 정렬
sorted_employees = sorted(employees, key=lambda e: e["salary"], reverse=True)

# 상위 3명 추출
top3 = [(emp["name"], emp["salary"]) for emp in sorted_employees[:3]]

print(top3)
```
