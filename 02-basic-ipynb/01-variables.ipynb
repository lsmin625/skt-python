# 파이썬 자료구조

## 파이썬 변수 유형

| 변수 유형 | 설명 | 예시 |
|---------|------|------|
| int | 정수형 숫자 | `age = 25` |
| float | 실수형 숫자 | `pi = 3.14` |
| str | 문자열 | `name = "홍길동"` |
| bool | 불리언 (참/거짓) | `is_active = True` |


````python
# 파이썬 헬로우 월드
print("Hello, World!")
````

````python
# 정수와 실수 예시
a = 10        # 정수
b = 3.14      # 실수
print("정수 a:", a)
print("실수 b:", b)
````

### 문자열
파이썬에서 문자열은 문자의 시퀀스로, 작은따옴표(') 또는 큰따옴표(")로 감쌀 수 있으며, 다양한 메서드를 사용하여 조작할 수 있다.

f-string(f 포맷터)는 문자열 내에서 변수나 표현식을 간단히 삽입할 수 있는 방법을 제공

```python
name = "Alice"
greeting = f"Hello, {name}!"
```

````python
# 문자열 예시
name = "Alice"
greeting = f"Hello, {name}!"
print(greeting)
````

### 불리언(True, False)
파이썬에서 불리언(Boolean)은 참(True)과 거짓(False)을 나타내는 데이터 유형이다. 조건문에서 주로 사용되며, 논리 연산의 결과로 반환된다.

```python
is_active = True
is_admin = False
print("활성 상태:", is_active)
print("관리자 여부:", is_admin)
```

````python
is_active = True
is_admin = False
print("활성 상태:", is_active)
print("관리자 여부:", is_admin)
````
