# 문자열

## 문자열 슬라이싱
문자열 슬라이싱은 문자열의 특정 부분을 추출하는 방법
- 슬라이싱 구문은 `문자열[시작:끝:단계]` 형식을 따르며, 시작 인덱스부터 끝 인덱스 전까지의 부분 문자열을 반환
- 단계는 선택적으로 지정할 수 있으며, 기본값은 1.

```python
# 문자열 슬라이싱 예시
text = "Hello, Python!"
substring = text[0:5]  # 'Hello'
print(f"슬라이싱된 문자열: {substring}")

# 문자열 뒤집기
reversed_text = text[::-1]
print(f"뒤집힌 문자열: {reversed_text}")
```

## 문자열 포맷팅

f-string은 문자열 앞에 `f` 또는 `F`를 붙여, 중괄호 `{}` 안에 변수나 표현식을 직접 삽입해 포맷팅하는 방법

```python
# f-string 예시
name = "Alice"
age = 30
greeting = f"{name}는 {age}살입니다."
print(greeting)

# 문자열 포맷팅 활용 예시
pi = 3.14159
formatted_pi = f"원주율은 {pi:.2f}입니다."
print(formatted_pi)

# 문자열 공백을 넣어 좌정렬 또는 우정렬
left_aligned = f"[{name:<10}]는 [{age:<5}]살입니다."
right_aligned = f"[{name:>10}]는 [{age:>5}]살입니다."
print(left_aligned)
print(right_aligned)
```
