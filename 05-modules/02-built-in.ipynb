# 파이썬 내장 모듈

## 파일 및 디렉토리 관리 - `os`

```python
# os 모듈 예시
import os

print(os.getcwd())  # 현재 작업 디렉토리 출력
print(os.listdir())  # 현재 디렉토리의 파일 목록 출력
```

## 무작위 요소 선택 - `random`

```python
# random 모듈 예시
import random

print(random.randint(1, 10))  # 1부터 10 사이의 무작위 정수 출력
print(random.choice(['apple', 'banana', 'cherry', '수박', '포도', '키위']))  # 무작위 과일 선택 출력

# shuffle 예시
items = [1, 2, 3, 4, 5]
random.shuffle(items)
print(items)  # 섞인 리스트 출력
```

## 날짜와 시간 - `dataetime`

```python
# datetime 모듈 예시
import datetime

now = datetime.datetime.now()
print("현재 날짜와 시간:", now)
print("연도:", now.year)
print("월:", now.month)
print("일:", now.day)
print("시:", now.hour)
print("분:", now.minute)
print("초:", now.second)

# 날짜 포맷팅 예시
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print("포맷팅된 날짜와 시간:", formatted_date)
```

## 파이썬 인터프리터 - `sys`

```python
# sys 모듈 예시
import sys

print("파이썬 버전:", sys.version)
print("모듈 검색 경로:", sys.path)
print("현재 모듈 이름:", sys.modules[__name__])  # 현재 모듈 정보 출력
```

## JSON 데이터 처리 - `json`

```python
# json 모듈 예시
import json

data = {"name": "Alice", "age": 30, "city": "New York"}
json_string = json.dumps(data)  # 딕셔너리를 JSON 문자열로 변환
print("JSON 문자열:", json_string)

parsed_data = json.loads(json_string)  # JSON 문자열을 딕셔너리로 변환
print("파싱된 데이터:", parsed_data)
```

## 정규 표현식 - `re`

```python
# re 모듈 예시
import re

pattern = r'\b\w{3}\b'  # 정확히 3글자인 단어 찾기
text = "The cat sat on the mat and saw a dog."
matches = re.findall(pattern, text)
print("3글자인 단어들:", matches)  # ['cat', 'sat', 'the', 'mat', 'and', 'saw', 'dog']

substituted_text = re.sub(r'dog', 'fox', text)
print("치환된 문장:", substituted_text)  # 'The cat sat on the mat and saw a fox.'
```
