## 조건문 (if)

```python
# if-else 문 예제
num = 7
if num % 2 == 0:
    print(f"{num}은(는) 짝수입니다.")
else:
    print(f"{num}은(는) 홀수입니다.")
```

```python

# if-elif-else 문 예제
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f"점수: {score}, 학점: {grade}")
```

## 조건문 (`match-case`)

- Python은 전통적인 switch 문 대신 `match-case` 문을 사용
- `match`는 주어진 값과 패턴을 비교하고, 일치하는 `case` 블록을 실행
- `case`에는 리터럴 값, 변수 바인딩, 와일드카드(_) 등을 사용
- 기본값 처리는 `case _`로 수행

```python
# switch 문 예제 (파이썬의 match-case 문 사용)
def switch_example(value):
    match value:
        case 1:
            return "하나"
        case 2:
            return "둘"
        case _:
            return "많다"

result = switch_example(3)
print(f"Switch 결과: {result}")
```
