# 연산자

## 산술 연산자

```python
# 사칙연산 예제
a = 10
b = 3

# 덧셈
c = a + b
print(f"{a} + {b} = {c}")

```

```python

# 뺄셈
d = a - b
print(f"{a} - {b} = {d}")

# 곱셈
e = a * b
print(f"{a} * {b} = {e}")

# 나눗셈
f = a / b
print(f"{a} / {b} = {f}")

# 몫
g = a // b
print(f"{a} // {b} = {g}")

# 나머지
h = a % b
print(f"{a} % {b} = {h}")

# 거듭제곱
i = a ** b
print(f"{a} ** {b} = {i}")
```

## 비교 연산자

```python
# 비교 연산자 예제
x = 100000
y = 100000

print(f"{x} == {y} : {x == y}")  # 같음

print(id(x), id(y))

```

```python

print(f"{x} != {y} : {x != y}")  # 같지 않음
print(f"{x} < {y} : {x < y}")    # 작음
print(f"{x} <= {y} : {x <= y}")  # 작거나 같음
print(f"{x} > {y} : {x > y}")    # 큼
print(f"{x} >= {y} : {x >= y}")  # 크거나 같음
```

## 논리 연산자

```python
# 논리 연산자 예제
p = True
q = False
print(f"{p} and {q} : {p and q}")  # 논리 AND
print(f"{p} or {q} : {p or q}")    # 논리 OR
print(f"not {p} : {not p}")        # 논리 NOT
```
