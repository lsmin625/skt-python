# 반복문

## for-in 문

```python
# for 문 예제1
for i in range(5):
    print(f"반복 횟수: {i}")
```

```python

# for 문 예제2
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(f"과일: {fruit}")
```

## while 문

```python
# while 문 예제1
count = 0
while count < 5:
    print(f"반복 횟수: {count}")
    count += 1
    # count = count + 1
```

```python

# while 문 예제2
num = 1
while num <= 10:
    if num % 2 == 0:
        print(f"{num}은(는) 짝수입니다.")
    num += 1
```

## 반복문 제어 (`break`, `continue`)

```python
# break 문 예제
for i in range(10):
    if i == 5:
        print("반복문 종료")
        break
    print(f"현재 값: {i}")

print("반복문 종료 후 실행되는 코드")
```

```python
# continue 문 예제
for i in range(10):
    if i % 2 == 0:
        continue
    print(f"현재 값(홀수만 출력): {i}")
```
