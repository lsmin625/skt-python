# 파이썬 내장 함수

## 기본 함수

```python
# 파이썬 기본 내장 함수
print("Hello, World!")  # 출력 함수
print(type(123))        # 자료형 확인
print(id("Python"))     # 객체의 고유 ID 확인
print(len([1, 2, 3]))   # 길이 확인
```

## 숫자 관련 함수

```python
# 숫자 관련 함수
print(f'abs(-10) = {abs(-10)}')                         # 절대값 반환
print(f'pow(2, 3) = {pow(2, 3)}')                       # 거듭제곱 반환
print(f'round(3.14159, 2) = {round(3.14159, 2)}')       # 반올림
print(f'max(1, 5, 3) = {max(1, 5, 3)}')                 # 최대값 반환
print(f'min(1, 5, 3) = {min(1, 5, 3)}')                 # 최소값 반환
print(f'sum([1, 2, 3, 4, 5]) = {sum([1, 2, 3, 4, 5])}') # 합계 반환
```

## 문자열 관련 함수

```python
# 문자열 관련 함수
print(f'chr(65) = {chr(65)}')                           # 아스키 코드에 해당하는 문자 반환
print(f'ord("A") = {ord("A")}')                         # 문자의 아스키 코드 반환
```

## 형 변환 함수

```python
# 형 변환 함수 - 변수를 통해 변환 전후의 type 표시
a = 123
b = str(a)
print(f'str(123) = {b}, type = {type(b)}')               # 문자열로 변환

c = float(a)
print(f'float(123) = {c}, type = {type(c)}')               # 실수로 변환

d = int(c)
print(f'int(123.0) = {d}, type = {type(d)}')               # 정수로 변환
```

## 리스트 & 튜플 관련 함수

```python
# 리스트 & 튜플 관련 함수 - len(), sorted(), max(), min(), sum()
lst = [3, 1, 4, 1, 5]
tpl = (3, 1, 4, 1, 5)

print(f'len(lst) = {len(lst)}')                           # 리스트 길이
print(f'sorted(lst) = {sorted(lst)}')                     # 리스트 정렬
print(f'max(tpl) = {max(tpl)}')                           # 튜플 최대값
print(f'min(tpl) = {min(tpl)}')                           # 튜플 최소값
print(f'sum(lst) = {sum(lst)}')                           # 리스트 합계 반환
```

## 기타 함수

### `enumerate()` - 인덱스와 값을 튜플로 반환

```python
fruits = ['apple', 'banana', 'cherry']
enumerated_fruits = enumerate(fruits)

print(fruits)
print(enumerated_fruits)

for index, fruit in enumerated_fruits:
    print(f'Index: {index}, Fruit: {fruit}')
```

### `zip()` - 여러 리스트를 병렬로 묶기

```python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f'Name: {name}, Age: {age}')
```

### `map()` - 리스트의 각 요소에 함수 적용

```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(f'Squared: {squared}')
```


### `filter()` - 조건에 맞는 요소만 추출

```python
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f'Even Numbers: {even_numbers}')
```
