# 파이썬 내장 함수

## 기본 함수

```python
# 파이썬 기본 내장 함수
print("Hello, World!")  # 출력 함수
print(type(123))        # 자료형 확인
print(id("Python"))     # 객체의 고유 ID 확인
print(len([1, 2, 3]))   # 길이 확인
```

## 숫자 관련 함수

```python
# 숫자 관련 함수
print(f'abs(-10) = {abs(-10)}')                         # 절대값 반환
print(f'pow(2, 3) = {pow(2, 3)}')                       # 거듭제곱 반환
print(f'round(3.14159, 2) = {round(3.14159, 2)}')       # 반올림
print(f'max(1, 5, 3) = {max(1, 5, 3)}')                 # 최대값 반환
print(f'min(1, 5, 3) = {min(1, 5, 3)}')                 # 최소값 반환
print(f'sum([1, 2, 3, 4, 5]) = {sum([1, 2, 3, 4, 5])}') # 합계 반환
```
