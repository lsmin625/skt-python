# 람다 함수와 고차 함수

## 람다 함수

람다 함수(lambda function)는 이름 없이 간단하게 정의할 수 있는 익명 함수로 한 줄로 작성되며, 함수를 간결하게 표현할 때 사용  
예시: `lambda x: x + 1` (입력값 x에 1을 더하는 함수)

## 고차 함수

고차 함수(higher-order function)는 함수를 인자로 받거나, 함수를 반환하는 함수로 람다 함수와 함께 자주 사용.  
- `map()`은 리스트의 각 요소에 함수를 적용
- `filter()`는 조건에 맞는 요소만 걸러냄

```python
# 람다 함수 예시1 - 리스트를 정렬할 때
students = [("Alice", 25), ("Bob", 20), ("Charlie", 23)]

# 나이 기준으로 정렬
students.sort(key=lambda student: student[0])
print(students)  # 출력: [('Bob', 20), ('Charlie', 23), ('Alice', 25)]
```

```python
# 람다 함수 예시2 - 고차 함수 map()과 함께 사용
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # 출력: [1, 4, 9, 16, 25]
```

```python
# 람다 함수 예시3 - 고차 함수 filter()와 함께 사용
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # 출력: [2, 4]
```

```python
# 람다 함수 예시4 - 고차 함수 reduce()와 함께 사용

# 모듈 임포트 필요
from functools import reduce

prices = [100, 200, 300, 400, 500]
total_price = reduce(lambda x, y: x + y, prices)
print(total_price)  # 출력: 1500
```
