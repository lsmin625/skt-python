# 튜플과 딕셔너리 언패킹

## 튜플 언패킹

튜플 언패킹(Tuple Unpacking)은 튜플의 요소들을 여러 개의 변수에 한 번에 할당하는 기능

- 코드를 간결하게 작성
- 여러 값을 동시에 할당
- 변수 값을 쉽게 교환

```python
# 튜플 언팩킹
person = ("Alice", 30, "Engineer")
name, age, profession = person
print(f"Name: {name}, Age: {age}, Profession: {profession}")
```

```python

# *args 언팩킹
def show_tuple(*args):
    print(f"Name: {args[0]}, Age: {args[1]}, Profession: {args[2]}")

person = ("Alice", 30, "Engineer")
show_tuple(*person)
```

## 딕셔너리 언패킹

딕셔너리 언패킹(Dictionary Unpacking)은 딕셔너리의 키-값 쌍을 함수의 인자나 변수에 한 번에 할당하는 기능

- 코드를 더욱 직관적이고 간결하게 작성.
- 함수 호출 시 키워드 인자를 쉽게 전달.
- 여러 변수에 값을 동시에 할당.

```python
# 딕셔너리 언패킹
def show_dict(name, age, profession):
    print(f"Name: {name}, Age: {age}, Profession: {profession}")

person = {
    "name": "Alice",
    "age": 30,
    "profession": "Engineer"
}

show_dict(**person)
```
