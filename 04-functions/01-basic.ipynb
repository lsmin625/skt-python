## 매개 변수가 포함된 함수

```python
# 매개 변수가 포함된 함수
def add(a, b):
    """
    이 함수는 두 수를 더하는 기능을 합니다.
    args:
        a (int, float): 첫 번째 숫자
        b (int, float): 두 번째 숫자
    returns:
        int, float: 두 수의 합
    """
    return a + b

# 빼기 함수
def subtract(a, b):
    """
    이 함수는 두 수를 빼는 기능을 합니다.
    args:
        a (int, float): 첫 번째 숫자
        b (int, float): 두 번째 숫자
    returns:
        int, float: 두 수의 차
    """
    return a - b

# result = add(3, 5)
result = subtract(10, 4)
print(f"결과: {result}")
```

## 매개 변수의 초기값이 정의된 함수

```python
# 매개 변수의 초기값이 정의된 함수
def power(base, exponent=2):
    """
    이 함수는 base의 exponent 제곱을 반환합니다.
    args:
        base (int, float): 밑수
        exponent (int): 지수 (기본값은 2)
    returns:
        int, float: base의 exponent 제곱
    """
    return base ** exponent 

result1 = power(4)        # exponent는 기본값 2 사용
print(f"4의 제곱: {result1}")

result2 = power(4, 3)     # exponent를 3으로 지정
print(f"4의 세제곱: {result2}")
```
