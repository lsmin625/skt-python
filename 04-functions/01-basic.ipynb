# 함수

## 함수 정의

```python
# 함수 정의 및 호출 예제1
def greet(name):
    print(f"안녕하세요, {name}님!")

greet("홍길동")
```

## 매개 변수가 포함된 함수

```python
# 매개 변수가 포함된 함수
def add(a, b):
    return a + b

result = add(3, 5)
print(f"결과: {result}")
```

## 매개 변수의 초기값이 정의된 함수

```python
# 매개 변수의 초기값이 정의된 함수
def power(base, exponent=2):
    return base ** exponent 

result1 = power(4)        # exponent는 기본값 2 사용
print(f"4의 제곱: {result1}")

result2 = power(4, 3)     # exponent를 3으로 지정
print(f"4의 세제곱: {result2}")
```
